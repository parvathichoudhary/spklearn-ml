```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randint(0,10,(1000,8)))
df.columns = ['cust_id','Gender','Age','Store','Unit','Transaction','sales','New/Existing']
```

```python
df['cust_id'] = ["cust_id"+str(i) for i in range(0,1000)]
df['Gender']=pd.DataFrame(np.random.randint(0,2,1000))
x = [] 
for i in df['Gender']:
    if i==0:
        x.append("Male")
    else:
        x.append("Female")
df['Gender']=x
df['Age']=pd.DataFrame(np.random.randint(10,100,1000))
x = [] 
for i in df['Age']:
    if i<18:
        x.append("0-18")
    elif i>=18 and i<24:
        x.append("18-14")
    elif i>=24 and i<35:
        x.append("24-35")
    elif i>=35 and i<50:
        x.append("35-50")
    else:
        x.append(">50")
df['Age']=x
str_num = np.random.randint(1,16,1000)
df['Store'] = ["str"+str(i) for i in str_num]
#df['Store'] = pd.DataFrame(np.random.randint(1,16,1000))
df['Transaction'] = pd.DataFrame(np.random.randint(1,16,1000))
df['sales'] = pd.DataFrame(np.random.randint(100,200,1000))
df['New/Existing'] = pd.DataFrame(np.random.randint(0,2,1000))
x = [] 
for i in df['New/Existing']:
    if i==0:
        x.append("New")
    else:
        x.append("Existing")
df['New/Existing']=x
df
#str_num
```
